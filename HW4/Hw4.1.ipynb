```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# 1. Створення вектору часу
speed = [25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65]
time = np.linspace(0, 11, num=12)
```

```python
# 2. Виведення вектору часу
print("Time vector:")
print(time)
```

```python
# 3. Графік швидкості
plt.figure(figsize=(10, 6))
plt.plot(time, speed, 'bo', label='Швидкість')
plt.xlabel('Час (години)')
plt.ylabel('Швидкість (км/год)')
plt.title('Графік швидкості')
plt.grid(True)
plt.xlim(0, 11)
plt.ylim(0, 130)
plt.legend()
plt.show()
```

```python
# 4. Інтерполяція та побудова безперервного графіку
interp_cubic = interp1d(time, speed, kind='cubic')
time_interp = np.linspace(0, 11, num=10000)
speed_interp = interp_cubic(time_interp)

plt.figure(figsize=(10, 6))
plt.plot(time_interp, speed_interp, 'r-', label='Інтерполяція')
plt.xlabel('Час (години)')
plt.ylabel('Швидкість (км/год)')
plt.title('Інтерполяція швидкості')
plt.grid(True)
plt.xlim(0, 11)
plt.ylim(0, 130)
plt.legend()
plt.show()
```

```python
# 5. Обчислення інтегралу для інтерполяційної функції (cubic)
integral_cubic = np.trapz(speed_interp, time_interp)
print("Інтеграл за допомогою cubic інтерполяції: ", integral_cubic)
```

```python
# Повторення для kind='quadratic'
interp_quadratic = interp1d(time, speed, kind='quadratic')
speed_interp_quadratic = interp_quadratic(time_interp)

plt.figure(figsize=(10, 6))
plt.plot(time_interp, speed_interp_quadratic, 'g-', label='Інтерполяція (quadratic)')
plt.xlabel('Час (години)')
plt.ylabel('Швидкість (км/год)')
plt.title('Інтерполяція швидкості (quadratic)')
plt.grid(True)
plt.xlim(0, 11)
plt.ylim(0, 130)
plt.legend()
plt.show()

integral_quadratic = np.trapz(speed_interp_quadratic, time_interp)
print("Інтеграл за допомогою quadratic інтерполяції: ", integral_quadratic)
```
