```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ввід даних
data = {'A': [25, 10, 10, 5, 35, 13],
        'B': [0, 15, -5, 5, 20, 25],
        'C': [10, 25, -15, -5, -5, 15]}
df = pd.DataFrame(data)

# обчислення коефіцієнтів кореляції
print(df.corr())
```

```python
# побудова ліній регресії та виведення графіків
sns.regplot(x='A', y='B', data=df)
plt.show()
```

```python
sns.regplot(x='B', y='C', data=df)
plt.show()
```

```python
sns.regplot(x='C', y='A', data=df)
plt.show()
```

```python
# створення фігури з трьома підграфіками
fig, axs = plt.subplots(ncols=3, figsize=(12, 4))

# побудова ліній регресії та виведення графіків
sns.regplot(x='A', y='B', data=df, ax=axs[0])
sns.regplot(x='B', y='C', data=df, ax=axs[1])
sns.regplot(x='C', y='A', data=df, ax=axs[2])

# встановлення заголовків для підграфіків
axs[0].set_title('A vs. B')
axs[1].set_title('B vs. C')
axs[2].set_title('C vs. A')

# відображення фігури з графіками
plt.tight_layout()
plt.show()
```
