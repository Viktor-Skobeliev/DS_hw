```python
import matplotlib.pyplot as plt
import random

# Кількість точок, які будуть згенеровані
n_points = 100000

# Вихідні точки трикутника
points = [[0, 0], [1, 0], [0.5, 0.866]]

# Зберігаємо координати всіх точок
x = []
y = []

# Вибираємо випадкову точку всередині трикутника
current_point = [random.uniform(0, 1), random.uniform(0, 1)]

for i in range(n_points):
    # Вибираємо випадкову вершину
    vertex = random.randint(0, 2)
    # Знаходимо точку, рівновіддалену від поточної точки та вибраної вершини
    current_point = [
        (current_point[0] + points[vertex][0]) / 2,
        (current_point[1] + points[vertex][1]) / 2,
    ]
    # Зберігаємо координати створеної точки
    x.append(current_point[0])
    y.append(current_point[1])

# Побудова візерунка
plt.scatter(x, y, s=1, c="black")
plt.axis("equal")
plt.axis("off")
plt.show()
```
